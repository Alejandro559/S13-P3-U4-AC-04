```python
#Paso 1: Carga de Datos Climáticos
import plotly.express as px

# Cargar el dataset integrado
wind_df = px.data.wind()

# Inspección inicial
print(wind_df.head())
print(wind_df.info())
```

```python
# Paso 2: Gráfico de Dispersión Polar (Scatter Polar)
fig1 = px.scatter_polar(wind_df, 
                        r="strength", 
                        theta="direction", 
                        color="strength",
                        symbol="direction",
                        size="frequency",
                        title="Dispersión Polar: Fuerza vs Dirección del Viento")
fig1.show()
```

```python
#Paso 3: Gráfico de Líneas Polar (Line Polar)
fig2 = px.line_polar(wind_df, 
                      r="strength", 
                      theta="direction", 
                      color="strength",
                      line_close=True, # Cierra el círculo de la línea
                      title="Tendencia Polar: Intensidad por Dirección")
fig2.show()
```

```python
#Paso 4: Gráfico de Barras Polar (Wind Rose)
fig3 = px.bar_polar(wind_df, 
                    r="frequency", 
                    theta="direction", 
                    color="strength",
                    color_discrete_sequence=px.colors.sequential.Plasma_r,
                    title="Rosa de los Vientos: Frecuencia por Dirección")
fig3.show()
```

```python
# Ejemplo de personalización del diseño
fig3.update_layout(
    font_size=12,
    legend_font_size=10,
    polar_angularaxis_rotation=90, # Rota el gráfico 90 grados
)

# Guardar como HTML
fig1.write_html("polar_scatter.html")
fig2.write_html("polar_line.html")
fig3.write_html("polar_bar.html")
```

1. ¿Qué patrones de dirección del viento predominantes observas en el gráfico de barras polar?
A partir de la Rosa de los Vientos (gráfico de barras), se observa un patrón predominante desde el Oeste (W), Oeste-Noroeste (WNW) y Noroeste (NW). Estas direcciones no solo son las más frecuentes (presentan las barras más largas), sino que también concentran las ráfagas de mayor intensidad, representadas por los niveles superiores de la escala de colores (tonos púrpura y rojo oscuro). Por el contrario, las direcciones Este (E) y Sur (S) muestran una actividad mínima.

2. ¿Cuál de las tres representaciones (dispersión, línea o barra) consideras que comunica mejor la intensidad del viento y por qué?
El gráfico de barras polar (Wind Rose) es el más efectivo. A diferencia del de dispersión, donde los puntos pueden solaparse y dificultar la lectura de la densidad de los datos, el gráfico de barras utiliza el apilamiento (stacking). Esto permite visualizar simultáneamente la frecuencia de cada dirección y la distribución de las distintas intensidades (fuerza) dentro de esa misma dirección mediante segmentos de color claramente diferenciados.

3. ¿Por qué es necesario usar un sistema de coordenadas polares en lugar de un gráfico de barras tradicional (rectangular) para datos de viento?
Es necesario debido a la naturaleza circular y continua de las direcciones. En un gráfico rectangular, el Norte aparecería dividido en dos extremos  lo que rompe la continuidad visual. El sistema polar imita una brújula, permitiendo que el usuario identifique la procedencia del viento de forma intuitiva y espacialmente correcta, manteniendo la relación lógica entre puntos cardinales adyacentes.
